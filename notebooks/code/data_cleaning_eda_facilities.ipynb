```python
# Import Libraries
import pandas as pd
```

Loading dataframes with VA facilities and facility types

```python
fac_df = pd.read_excel('../data/raw_data/va_facilities_list.xlsx')
fac_type_df = pd.read_csv('../data/raw_data/facility_type_list.csv')
```

```python
fac_type_df.info()
```

Cleaning fac_df

```python
fac_df.columns = ['id', 'name', 'address', 'state', 'phone']
fac_df.sort_values(by='id', inplace=True)
fac_df.head()
```

```python
fac_df.info()
```

```python
fac_df.isna().sum()
```

```python
fac_df[fac_df['id'].isna()]
```

```python
fac_df.dropna(inplace=True)
fac_df.drop(columns='phone', inplace=True)
fac_df.info()
```

```python
fac_df['state'].unique()
```

```python
not_state = ['PR', 'VI', 'AS', 'GU', 'MP', 'PI']
fac_df = fac_df[fac_df['state'].isin(not_state) == False]
```

```python
#fac_df = fac_df[fac_df.duplicated(['id']) == False]
```

Cleaning fac_type_df

```python
fac_type_df.info()
```

```python
fac_type_df.columns = ['id', 'location', 'state', 'zip', 'visn', 'type', 'designation']
fac_type_df.sort_values(by='id', inplace=True)
```

```python
fac_type_df['state'].unique()
```

```python
fac_type_df = fac_type_df[fac_type_df['state'].isin(not_state) == False]
```

```python
fac_type_df['state'].unique()
```

```python
fac_type_df.head()
```

```python
fac_type_df = fac_type_df[['id', 'type']]
```

Merging fac_df and fac_type_df

```python
facilities = fac_df.merge(fac_type_df, on='id', how='left')
facilities[facilities['id'] == '757']
```

```python
facilities.info()
```

```python
facilities[facilities['name'].str.contains('Clermont')]
```

```python
facilities.loc[facilities['name'].str.contains('CBOC'), 'type'] = 'CBOC'
facilities.loc[facilities['name'].str.contains('Community Based Outpatient Clinic'), 'type'] = 'CBOC'
facilities.loc[facilities['name'].str.contains('Vet Center'), 'type'] = 'VC'
```

```python
facilities.isna().sum()
```

```python
facilities['type'].fillna('OOS', inplace=True)
```
