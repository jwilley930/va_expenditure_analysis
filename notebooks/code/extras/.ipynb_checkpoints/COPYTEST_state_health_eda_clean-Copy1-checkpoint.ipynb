```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
state_health = pd.read_csv('../data/raw_data/health_info_by_state.csv', index_col=0)
```

```python
state_health.head()
```

```python
state_health = state_health.T
```

```python
state_health.head()
```

```python
#list(state_health.columns)
```

```python
# Making column names readable
health_dict = {'index': 'state',
'Total Residents, 2019': 'total_pop',
 'Death Rate per 100,000, 2020': 'deaths_per_100k',
 'Age adjusted Cancer Incidence per 100,000, 2018': 'cancer_per_100k',
 'Cancer Death Rate per 100,000, 2020':'cancer_death_per_100k',
 'Adults who have been told they have diabetes, 2020': 'diabetes',
 'Adults who have been told they have pre-diabetes, 2020': 'prediabetes',
 'Diabetes Death Rate per 100,000 total population, 2020': 'diabetes_death_per_100k',
 'Adults Who Report Being Told by a Doctor that They Have Cardiovascular Disease by Sex, 2020': 'cardio_disease',
 'Adults Who Have Ever Been Told by a Doctor that They Have Hypertension, 2019': 'hypertension',
 'Heart Disease Death Rate per 100,000 total population, 2020': 'heart_disease_death_per_100k',
 'Adult Overweight/Obesity Rate, 2017': 'overweight_obese',
 'Adults Who Are Severely Obese(BMI 40 or higher), 2020': 'severely_obese',
 'Asthma Prevalence Among Adults, 2017': 'asthma',
 'Suicide Rate per 100,000 Individuals, 2020': 'suicide_per_100k',
 'Adults Reporting Any Mental Illness in the Past Year, 2020': 'mental_illness',
 'Adults Reporting Serious Mental Illness in the Past Year, 2020': 'serious_mental_illness',
 'Percent of Adults Who Smoke, 2017': 'smoker',
 'Adults Reporting Alcohol Dependence or Abuse in the Past Year, 2019-2020': 'alcohol_abuse',
 'Alcohol-Induced Deaths per 100,000, 2020': 'alcohol_death_per_100k',
 'Adults Reporting Illicit Drug Dependence or Abuse in the Past Year, 2019-2020': 'drug_abuse',
 'Drug Overdose Death Rate (per 100,000 population), 2020': 'drug_death_per_100k',
 'Opioid Overdose Death Rate per 100,000 Population (Age-Adjusted), 2020': 'opiod_death_per_100k',
 'Adults Who Report Participation in Any Physical Activity or Exercise, 2020': 'exercise',
 'Adults Reporting Not Having a Personal Doctor, 2020': 'no_doctor',
 'Adults Who Self-Report Fair or Poor Health Status': 'self_report_low_health'}

state_health.rename(columns=health_dict, inplace=True)
state_health.head()
```

```python
state_health.info()
```

```python
#Converting data to numeric
state_health.replace(',', '', regex=True, inplace=True)
state_health.replace('%', '', regex=True, inplace=True)

state_health = state_health.astype('float')
```

```python
# Converting all values to floats representing proportions of total
per_100k_cols = ['deaths_per_100k', 'cancer_per_100k', 'cancer_death_per_100k', 'diabetes_death_per_100k', 'heart_disease_death_per_100k', 'suicide_per_100k', 'alcohol_death_per_100k', 'drug_death_per_100k', 'opiod_death_per_100k']
percent_cols = ['state', 'diabetes', 'prediabetes', 'cardio_disease', 'hypertension', 'overweight_obese', 'severely_obese', 'asthma', 'mental_illness', 'serious_mental_illness', 'smoker', 'alcohol_abuse', 'drug_abuse', 'exercise', 'no_doctor', 'self_report_low_health']

state_health[per_100k_cols] = state_health[per_100k_cols]/100000
state_health[percent_cols] = state_health[percent_cols]/100
```

### TESTING

```python
state_health = state_health[percent_cols]
```

```python
#Filling two missing values with column means
state_health = state_health.fillna(state_health.mean())
```

```python
state_health.isna().sum()
```

```python
state_health.reset_index(inplace=True)
state_health.rename(columns={'index': 'state'}, inplace=True)
state_health.head()
```

```python
state_health.to_csv('../data/state_health.csv', index=False)
```
