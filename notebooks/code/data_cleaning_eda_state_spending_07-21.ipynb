```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# loading all sheets of the file into single dataframe, adding the sheet name as the year column
# referenced to add sheet_name as column:  https://stackoverflow.com/questions/59328721/how-do-i-add-a-column-to-my-dataframe-that-says-what-sheet-name-each-row-is-from
file = pd.ExcelFile('../data/raw_data/AGG_GDX_FY07-21.xlsx')
years = file.sheet_names

exp = pd.concat([pd.read_excel(file, sheet_name=s).assign(year=s) for s in years])
exp.head()
```

```python
# dropping column, cleaning column names
exp.drop(columns='Loan Guaranty#', inplace=True)
exp.columns = ['state', 'vet_pop', 'total_exp', 'comp_&_pension', 'construction', 'edu_training', 'op_exp', 'insurance', 'medical_care', 'unique_patients', 'year']
exp.head()
```

```python
exp.set_index('year').head()
```

```python
# moving year column to front
year_col = exp.pop('year')
exp.insert(1, year_col.name, year_col.astype(int))
```

```python
exp.sort_values(by='year')
exp.sort_values(by=['state','year'], inplace=True)
```

```python
exp.head()
```

```python
alabama = exp[exp['state']=='Alabama']
```

```python
arkansas = exp[exp['state']=='Arkansas']
```

```python
plt.plot(alabama['year'], alabama['total_exp'])
plt.plot(arkansas['year'], arkansas['total_exp'])
```
